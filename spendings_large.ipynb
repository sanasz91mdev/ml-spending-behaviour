```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

mccs = pd.read_csv('./Data_2000_SA.csv')
mccs['Target-MCC1'] = mccs['Target-MCC1'].astype('category')
mccs.head()

# Define the dictionary with the relabeling
label_dict = {
    7832: '7832 - Motion Picture Theaters',
    5411: '5411 - Grocery Stores', 
    5812: '5812 - Eating places and Restaurant',
    7032: '7032 - Sporting and Recreational Camps',
    5999: '5999 - Miscellaneous and Specialty Retail Stores',
    4111: '4111 - Transportation',
    5691: '5691 - Men and Women Clothing Stores',
    5983: '5983 - Fuel – Fuel Oil',
    4814: '4814 - Telecommunication services', 
    5621: '5621 - Women Ready-to-Wear Stores'
}

# Apply the relabeling to the column in the data frame
mccs['Target-MCC1'] = mccs['Target-MCC1'].map(label_dict)



# # Create a bar plot showing the count of each Target-MCC1 category for each Gender category
sns.countplot(x='Gender', hue='Target-MCC1', data=mccs)

# Set the title and axes labels
plt.title('Relationship between Gender and Target-MCC1')
plt.xlabel('Gender')
plt.ylabel('Count')


# Show the plot
plt.show()
```

```python

# Create a stacked bar plot showing the count of each Target-MCC1 category for each combination of AgeGroup, Gender, and MaritalStatus
sns.catplot(x='AgeGroup', hue='Target-MCC1', col='MaritalStatus', row='Gender', data=mccs, kind='count', height=4, aspect=1)


# Set the title and axes labels
plt.suptitle('Relationship between AgeGroup, Gender, and MaritalStatus with Target-MCC1')
plt.subplots_adjust(top=0.9)
plt.xlabel('AgeGroup')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python

# Create a stacked bar plot showing the count of each Target-MCC1 category for each combination of Day, Gender, and MaritalStatus
sns.catplot(x='Day', hue='Target-MCC1', col='MaritalStatus', row='Gender', data=mccs, kind='count', height=4, aspect=1)


# Set the title and axes labels
plt.suptitle('Relationship between Day, Gender, and AgeGroup with Target-MCC1')
plt.subplots_adjust(top=0.9)
plt.xlabel('Day')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python

# Create a stacked bar plot showing the count of each Target-MCC1 category for each combination of AgeGroup, Gender, and MaritalStatus
sns.catplot(x='City', hue='Target-MCC1', col='MaritalStatus', row='Gender', data=mccs, kind='count', height=4, aspect=1)


# Set the title and axes labels
plt.suptitle('Relationship between City, Gender, and MaritalStatus with Target-MCC1')
plt.subplots_adjust(top=0.9)
plt.xlabel('AgeGroup')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
corr = mccs.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Plot the heatmap
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0, annot=True, fmt=".2f")

# Show the plot
plt.show()
```
