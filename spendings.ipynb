```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

mccs = pd.read_csv('./data-small-more-patterns.csv')
mccs['Target-MCC1'] = mccs['Target-MCC1'].astype('category')
mccs.head()
```

```python
grouped_data = mccs.groupby(['AgeGroup', 'Target-MCC1']).mean().reset_index()
sns.barplot(x='AgeGroup', y='Avg7032', hue='Target-MCC1', data=grouped_data)
plt.show()
```

```python
# Create a bar plot showing the count of each Target-MCC1 category for each AgeGroup category
sns.countplot(x='AgeGroup', hue='Target-MCC1', data=mccs)

# Set the title and axes labels
plt.title('Relationship between AgeGroup and Target-MCC1')
plt.xlabel('AgeGroup')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Create a bar plot showing the count of each Target-MCC1 category for each Gender category
sns.countplot(x='Gender', hue='Target-MCC1', data=mccs)

# Set the title and axes labels
plt.title('Relationship between Gender and Target-MCC1')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python

# Create a stacked bar plot showing the count of each Target-MCC1 category for each combination of AgeGroup, Gender, and MaritalStatus
sns.catplot(x='AgeGroup', hue='Target-MCC1', col='MaritalStatus', row='Gender', data=mccs, kind='count', height=4, aspect=1)


# Set the title and axes labels
plt.suptitle('Relationship between AgeGroup, Gender, and MaritalStatus with Target-MCC1')
plt.subplots_adjust(top=0.9)
plt.xlabel('AgeGroup')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python

# Create a stacked bar plot showing the count of each Target-MCC1 category for each combination of AgeGroup, Gender, and MaritalStatus
sns.catplot(x='City', hue='Target-MCC1', col='MaritalStatus', row='Gender', data=mccs, kind='count', height=4, aspect=1)


# Set the title and axes labels
plt.suptitle('Relationship between City, Gender, and MaritalStatus with Target-MCC1')
plt.subplots_adjust(top=0.9)
plt.xlabel('AgeGroup')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
corr = mccs.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Plot the heatmap
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0, annot=True, fmt=".2f")

# Show the plot
plt.show()
```
